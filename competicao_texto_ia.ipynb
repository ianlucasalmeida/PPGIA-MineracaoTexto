## Pré-processamento dos dados

### Subtask:
Limpar e preparar os dados de texto para o treinamento do modelo (por exemplo, tokenização, remoção de stop words, lematização/stemming).


**Reasoning**:
Define a function to preprocess the text data and apply it to the text columns of the training and testing dataframes.



```python
import re
import nltk
from nltk.corpus import stopwords

# Download necessary NLTK data
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')  # <- necessário agora
nltk.download('rslp') # Portuguese stemmer

# Initialize the Portuguese stemmer
stemmer = nltk.stem.RSLPStemmer()

# Get Portuguese stop words
stop_words = set(stopwords.words('portuguese'))

def preprocess_text(text):
    """
    Preprocesses the input text: converts to lowercase, removes punctuation,
    removes stop words, and applies stemming.
    """
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenize the text
    tokens = nltk.word_tokenize(text, language="portuguese")
    # Remove stop words and apply stemming
    preprocessed_tokens = [stemmer.stem(word) for word in tokens if word not in stop_words]
    # Join tokens back into a string
    return ' '.join(preprocessed_tokens)

# Apply the preprocessing function to the 'text' column
df_train['preprocessed_text'] = df_train['text'].apply(preprocess_text)
df_test['preprocessed_text'] = df_test['text'].apply(preprocess_text)

# Display the first few rows with the new column
print("Training Data with Preprocessed Text:")
display(df_train.head())

print("\nTest Data with Preprocessed Text:")
display(df_test.head())
```

```python
import re
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Load the datasets
train_file_path = '/content/kaggle_train.csv'
test_file_path = '/content/kaggle_test.csv'

try:
    df_train = pd.read_csv(train_file_path)
    df_test = pd.read_csv(test_file_path)
    print("Data loaded successfully.")
except FileNotFoundError as e:
    print(f"Error loading data: {e}")
    # Exit or handle the error appropriately if files are not found
    exit() # Or sys.exit() if sys is imported

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('punkt_tab') # Explicitly download punkt_tab
nltk.download('stopwords')
nltk.download('rslp') # Portuguese stemmer

# Initialize the Porter Stemmer for Portuguese
stemmer = nltk.stem.RSLPStemmer()

# Get Portuguese stop words
stop_words = set(stopwords.words('portuguese'))

def preprocess_text(text):
    """
    Preprocesses the input text: converts to lowercase, removes punctuation,
    removes stop words, and applies stemming.
    """
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenize the text
    tokens = nltk.word_tokenize(text)
    # Remove stop words and apply stemming
    preprocessed_tokens = [stemmer.stem(word) for word in tokens if word not in stop_words]
    # Join tokens back into a string
    return ' '.join(preprocessed_tokens)

# Apply the preprocessing function to the 'text' column
df_train['preprocessed_text'] = df_train['text'].apply(preprocess_text)
df_test['preprocessed_text'] = df_test['text'].apply(preprocess_text)

# Display the first few rows with the new column
print("Training Data with Preprocessed Text:")
display(df_train.head())

print("\nTest Data with Preprocessed Text:")
display(df_test.head())
```

## Engenharia de features

### Subtask:
Transformar os dados de texto pré-processados em representações numéricas que podem ser usadas pelos modelos de machine learning (por exemplo, TF-IDF ou embeddings de palavras).


**Reasoning**:
Transform the preprocessed text data into numerical representations using TF-IDF.



```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Instantiate TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform on training data
X_train_tfidf = tfidf_vectorizer.fit_transform(df_train['preprocessed_text'])

# Transform on test data
X_test_tfidf = tfidf_vectorizer.transform(df_test['preprocessed_text'])

# Print the shape of the resulting TF-IDF matrices
print("Shape of X_train_tfidf:", X_train_tfidf.shape)
print("Shape of X_test_tfidf:", X_test_tfidf.shape)
```

## Treinamento do modelo

### Subtask:
Selecionar e treinar um ou mais modelos de classificação de texto nos dados de treino com as features criadas.


**Reasoning**:
Import the Logistic Regression model and train it on the TF-IDF transformed training data.



```python
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
model = LogisticRegression()

# Train the model
model.fit(X_train_tfidf, df_train['label'])

print("Model training complete.")
```
